This Python code analyzes the frequency of each alphabet character in a given text and visualizes the results in a horizontal bar plot using `matplotlib`.

- **Plotting**:

```python
# Create and activate a Python virtual environment. 
# Before starting the project, try all these imports FIRST
# Address any errors you get running this code cell 
# by installing the necessary packages into your active Python environment.
# Try to resolve issues using your materials and the web.
# If that doesn't work, ask for help in the discussion forums.
# You can't complete the exercises until you import these - start early! 
# We also import Counter and randint (included in the Python Standard Library).

from collections import Counter
from random import randint
import matplotlib.pyplot as plt

print('All prereqs installed.')
!pip list


from collections import Counter
TEXT = "At three o’clock precisely I was at Baker Street, but Holmes had not yet returned. The landlady informed me that he had left the house shortly after eight o’clock in the morning. I sat down beside the fire, however, with the intention of awaiting him, however long he might be. I was already deeply interested in his inquiry, for, though it was surrounded by none of the grim and strange features which were associated with the two crimes which I have already recorded, still, the nature of the case and the exalted station of his client gave it a character of its own. Indeed, apart from the nature of the investigation which my friend had on hand, there was something in his masterly grasp of a situation, and his keen, incisive reasoning, which made it a pleasure to me to study his system of work, and to follow the quick, subtle methods by which he disentangled the most inextricable mysteries. So accustomed was I to his invariable success that the very possibility of his failing had ceased to enter into my head.".lower()
letter_freq = dict(Counter(TEXT))

# make a bar plot with an appropriate title, correct xtick labels, and labeled axes

# Filter out non-alphabet characters for clarity in the plot
letter_freq = {k: v for k, v in letter_freq.items() if k.isalpha()}

# Sort the dictionary alphabetically by key
letter_freq = dict(sorted(letter_freq.items()))

# Create a horizontal bar plot
plt.figure(figsize=(10, 8))
plt.barh(list(letter_freq.keys()), list(letter_freq.values()), color='red')
plt.title("Frequency Distribution of Characters in Text")
plt.xlabel("Frequency")
plt.ylabel("Character")

# Show the plot
plt.show()
```

# Random Number Visualization with Scatter and Line Plots

This Python code generates a sequence of 10 random integers and displays them using both a scatter plot and a line plot. The `matplotlib` library is used for plotting, while the `random` library generates the random numbers.

- **Data Generation**: A list of 10 random integers between 1 and 100 is created for the y-values, with indices from 0 to 9 as the x-values.
- **Plotting Setup**: 
  - A figure is created with two subplots arranged side-by-side.
  - The left subplot displays a scatter plot, where each random number is represented by a blue dot.
  - The right subplot displays a line plot, where each random number is connected by a green line with markers at each point.
- **Customization**: 
  - Each plot includes titles, x-axis labels (`Index`), and y-axis labels (`Random Value`) for clarity.
- **Display**: The `tight_layout` function optimizes the layout before displaying the plots side by side.

This code provides a simple example of using scatter and line plots to visualize data.


```python
import matplotlib.pyplot as plt
from random import randint

# Generate a sequence of 10 random integers
y = [randint(1, 100) for _ in range(10)]
x = list(range(len(y)))

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Scatter plot
ax1.scatter(x, y, color='blue')
ax1.set_title("Scatter Plot of Random Numbers")
ax1.set_xlabel("Index")
ax1.set_ylabel("Random Value")

# Line plot
ax2.plot(x, y, color='green')
ax2.set_title("Line Plot of Random Numbers")
ax2.set_xlabel("Index")
ax2.set_ylabel("Random Value")

# Show the plots
plt.tight_layout()
plt.show()
```

# Enhanced Random Number Visualization with Customized Scatter and Line Plots

This Python code generates and visualizes a sequence of 10 random integers using both a scatter plot and a line plot. The visual style is enhanced for clarity and aesthetics using `matplotlib`.

- **Data Generation**: 
  - A list of 10 random integers between 1 and 100 is generated for the y-values, and the x-values are the indices (0–9).
- **Plotting Setup**: 
  - A figure with two subplots is created, arranged side-by-side.
  - The left subplot displays a scatter plot with customized markers:
    - **Color**: Purple squares with a black edge.
    - **Size**: Larger markers for better visibility.
    - **Transparency**: Set to 70% opacity.
  - The right subplot displays a line plot with custom styling:
    - **Color**: Orange with dashed lines and diamond-shaped markers.
    - **Line and Marker Styling**: Thicker line width, larger marker size, and 80% opacity.
- **Customization**: 
  - Both plots have titles, x-axis labels (`Index`), and y-axis labels (`Random Value`) for clear presentation.
- **Display**: 
  - `tight_layout` adjusts the layout for optimal viewing before displaying the plots.

This code demonstrates how to use custom styling in scatter and line plots to create visually appealing data visualizations.


```python
import matplotlib.pyplot as plt
from random import randint

# Generate a sequence of 10 random integers
y = [randint(1, 100) for _ in range(10)]
x = list(range(len(y)))

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Scatter plot
ax1.scatter(x, y, color='purple', marker='s', s=100, edgecolor='black', alpha=0.7)
ax1.set_title("Scatter Plot of Random Numbers")
ax1.set_xlabel("Index")
ax1.set_ylabel("Random Value")

# Line plot
ax2.plot(x, y, color='orange', marker='D', markersize=8, linestyle='--', linewidth=2, alpha=0.8)
ax2.set_title("Line Plot of Random Numbers")
ax2.set_xlabel("Index")
ax2.set_ylabel("Random Value")

# Show the plots
plt.tight_layout()
plt.show()
```

# Sorting Algorithm Performance Comparison: Insertion Sort vs. Merge Sort

This Python code compares the execution times of two sorting algorithms—Insertion Sort and Merge Sort—on lists of varying sizes, using `matplotlib` to plot the data.

- **Data**:
  - `sizes`: List of different input sizes for sorting, ranging from 10 to 5000.
  - `insertion_sort_times` and `merge_sort_times`: Lists representing the time (in seconds) each algorithm took to sort lists of each respective size.

- **Plotting**:
  - A single figure is created with two line plots:
    - **Insertion Sort**: Plotted in blue with circular markers and a solid line.
    - **Merge Sort**: Plotted in red with square markers and a dashed line.
  - **Labels**: Both the x-axis (`List Size`) and y-axis (`Time (seconds)`) are labeled to describe the data.
  - **Title**: The plot is titled "Execution Time for Sorting Algorithms by List Size" for clarity.

- **Legend and Grid**:
  - A legend distinguishes the two algorithms, and gridlines are added for better readability.

This visualization provides an easy comparison of the time complexity of Insertion Sort and Merge Sort as the list size increases.


```python
import matplotlib.pyplot as plt

sizes = [10, 20, 50, 100, 1000, 5000]
insertion_sort_times = [0.0010252999999999998, 0.0027026999999999954, 0.010147200000000002, 0.0381137, 3.6303399, 91.2180796]
merge_sort_times = [0.00161889999999687, 0.003635600000009731, 0.0090655000000055, 0.020108000000007564, 0.2687474000000094, 1.6147050999999948]


# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(sizes, insertion_sort_times, label='Insertion Sort', color='blue', marker='o', linestyle='-')
plt.plot(sizes, merge_sort_times, label='Merge Sort', color='red', marker='s', linestyle='--')

# Adding labels and title
plt.xlabel("List Size")
plt.ylabel("Time (seconds)")
plt.title("Execution Time for Sorting Algorithms by List Size")
plt.legend()  # Add legend to distinguish the sorting algorithms

# Show the plot with gridlines for better readability
plt.grid(True)
plt.show()
```
